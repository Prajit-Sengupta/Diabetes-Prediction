```python
import pandas as pd

# Reading the CSV file
df = pd.read_csv('C:/Users/anubh/OneDrive/Desktop/diabetes.csv')

# Display the first 5 rows of the dataframe
print(df.head())
```

```python
#Z-Score Normalization
parameters = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

for param in parameters:
    mean = df[param].mean()
    df[param] = df[param] - mean
    max_value = df[param].max()
    df[param] = df[param] / max_value

df
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# Create a figure and a set of subplots
fig, axes = plt.subplots(3, 3, figsize=(15, 15))

# List of columns to plot
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age']

# Plot each column
for i, column in enumerate(columns):
    sns.histplot(df[column], kde=True, ax=axes[i//3, i%3])
    axes[i//3, i%3].set_title(f'Distribution of {column}')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
# Loop through each parameter and plot the distribution for Outcome = 1
for param in parameters:
    feature_outcome_1 = df[df['Outcome'] == 1][[param, 'Outcome']]
    
    plt.figure(figsize=(10, 6))
    plt.hist(feature_outcome_1[param], bins=20, edgecolor='k', alpha=0.7)
    plt.title(f'{param} Distribution for Outcome = 1')
    plt.xlabel(param)
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

```
